```python
### Introduction to Python
# Variables, types and basic syntax

# First of all, you do not need characters like ';' to end your lines. Python uses whitespace for this.

# Note that in Python you do not need to specify types for your variables. They are inferred from data.
# Variables are declared and initialized as follows
x = 7 #Type: int
y = 3.0 #Type: float
z = "Hello World" #Type: string

# The print command is an easy way to get information out of your script
print(x)
print(y)
print(z)

# We can also use f strings for ease in formatting
print(f"\n\n{z}! Here's the result of calculating x/y: {x/y:.2f}, rounded to 2 decimals.\n\n")

# Type inference in python is an important point, so try to keep track of what your types are
ratio = x / y #What type does this have? What if x was equal to e.g. 12?
some_integer = 4
some_integer_or_is_it = 4.0

# You can use type() to get information about the type of an object 
print(type(ratio))
```

```python
# Control Structures

# if - elif - else
if(x > y): # Note that you can use "or", "and" and "not" as well as other logical operators to combine conditions.
    print("Condition met!")
elif(x==y):
    print("Alternative condition ment. This shouldn't be the case.")
else:
    print("... and this shouldn't be the case either.")

# while
while(x>y):
    x -= 1
    #x = x-1
    print(x)

# for - This is likely the most frequently used type of loop
for i in range(0,3):
    x = x**2 #Square x a modest, few times
    print(x)

x = 7 #Reset x to not overflow at some point
```

```python
# Data Types and Containers
# Python has all of the classic data types (int, float, string, etc.). Here we will look at various data storing structures.

# Array
py_array = [1,2,3,4,5]
print(py_array)
# Indexing beings at 0
py_array[0] = 6
print(py_array)
# You can use negative indexes to count from the back of the array
py_array[-1] = 7
print(py_array)
# You can also use the following notation to "slice" the array
print(py_array[2:4])
# There are also multidimensional arrays
multi_array = [[1,2,3], [4,5,6]]
print(multi_array[1][2])

# Dictionary
# In dictionaries you can combine a key with a value. They may of any type. You acces the dictionary by giving it a key and receiving a value.
py_dict = {"key": "value",
           "number" : 12345,
           3.14 : 42}

print(f"Value at key 'key': {py_dict["key"]}")
print(f"Value at key 'number': {py_dict["number"]}")
print(f"Value at key '3.14': {py_dict[3.14]}")

```

Python uses packages to enhance its functionality. We will be using a few select packages for numerical handling of large amounts of numbers, as well as packages for plotting and ML packages. There are rich documentations of these packages available online.

```python
# In the following we include two packages
import numpy as np # Numpy is used for many numerical functions and for easily handling large amounts of numbers
import matplotlib.pyplot as plt # The subpackage pylot of matplotlib is a power plotting tool

# Numpy offers its own class of array which can be initialized as follows
# They offer more functionality than python arrays
numpy_array = np.array([1,2,3])

# You can also quickly generate arrays, e.g. an all zero matrix of dimensions 2x4
all_zeros = np.zeros((2,4))
all_ones = np.ones((3,3))

print(f"The maximum of the array {numpy_array} is {numpy_array.max()}")

# Note, the shape property is useful for debugging
print(f"The zero matrix\n {all_zeros}\n has dimensions {all_zeros.shape}")

# Numpy supports matrix/vector operations
# Make sure to have the right dimensions!
print(f"Multiplying element-wise: {numpy_array * numpy_array}")
print(f"Matrix multiplication: {numpy_array @ all_ones}")

#There are various tricks for manipulating numpy arrays
example_matrix = np.array([
    [1,2,3],
    [4,5,4],
    [3,2,1]
])

print(f"Creating a matrix by repeating array rows:\n {np.tile(numpy_array, (3,1))}")
print(f"Calculating properties of matrices:\n Transpose:\n{example_matrix.T}\nInverse:\n{np.linalg.inv(example_matrix)}")
```

In order to visualize your results, you can use the pyplot module of the matplotlib package. As we have already imported it in the cell above, we may now use it.

```python
# This is a function definition. It takes in a value, calculates a cubic polynomial and returns the result.
def cubic_func(x):
    return 0.78 * x**3 - 12 *x**2 + 43 *x - 64  
```

```python
# We generate linearly spaced points between 0 and 12
x = np.linspace(0, 12, 100) 
# Then we calculated the results at the various points
# Note that we are passing a numpy array, not just a single number, numpy can pass entire arrays 
y = cubic_func(x)

# plt.figure creates a figure for our plots to live in
plt.figure(figsize=(10, 4))

#First subplot
plt.subplot(1, 2, 1)
# This is a quick and simple way to get a plot on the screen, e.g. for debugging
plt.plot(x, y)
plt.title("Simple Plot")

#Second subplot
plt.subplot(1, 2, 2)
# The following snippet showcases some more options to make the plot look nicer
plt.plot(x, y, 'r--', label='Cubic function')
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Cubic Function plot")
plt.grid(True)
plt.legend()
plt.xlim(0, 12)
plt.ylim(min(y) - 1, max(y) + 1)
plt.yscale('linear')  #'log' for logarithmic scaling



```
